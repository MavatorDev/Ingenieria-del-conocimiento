```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv('energydata.csv')
dframe=pd.DataFrame(datos)
x=dframe[['T8','T_out']]#entradas
y=dframe['RH_8']#
```

```python
#mostramos qué tenemos
print("Datos originales")
dframe.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.9)
print("X_train")
print(X_train)
print("y_train")
print(y_train)
```

```python
regresion=linear_model.LinearRegression()
regresion.fit(X_train,y_train)
ypredict=regresion.predict(X_test)#
```

```python
fig = plt.figure()
plt.scatter(X_train[0:60],y_train[0:60],color='blue')
plt.scatter(X_test[0:60],y_test[0:60],color='red')
fig.savefig('grafica.png')
```

```python
print("Error: ",mean_squared_error(y_test,ypredict))
print("El valor de r^2: ",r2_score(y_test,ypredict))
```
